Importar bibliotecas

```python
from gtts import gTTS
from IPython.display import Audio, display
```

Definir texto original (em português)

```python
texto_original = "Eu vou ao supermercado"
```

 Criar áudio em inglês

```python
audio_ingles = gTTS(text="I am going to the supermarket", lang="en", slow=False)
audio_ingles.save("/content/audio_ingles.wav")
```

Criar áudio em português

```python
audio_portugues = gTTS(text=texto_original, lang="pt", slow=False)
audio_portugues.save("/content/audio_portugues.wav")
```

Criar áudio em mandarim (chinês simplificado)

```python
audio_chines = gTTS(text="我要去超市", lang="zh-CN", slow=False)
audio_chines.save("/content/audio_chines.wav")
```

Tocar os áudios

```python
print("Inglês:")
display(Audio("/content/audio_ingles.wav"))

print("Espanhol:")
display(Audio("/content/audio_espanhol.wav"))

print("Mandarim:")
display(Audio("/content/audio_chines.wav"))
```
